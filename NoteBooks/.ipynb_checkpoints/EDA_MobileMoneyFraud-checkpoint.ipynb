## Investigate Balance Behavior

We want to find suspicious or illogical patterns in:

- `oldbalanceOrg`, `newbalanceOrig` ‚Üí For the sender
- `oldbalanceDest`, `newbalanceDest` ‚Üí For the receiver

Especially when:

- The transaction is marked as **fraud** (`isFraud == 1`)
- Or is **flagged** (`isFlaggedFraud == 1`)

### 1. Does the amount match the change in sender balance?

```python
df['errorOrig'] =df['oldbalanceOrg'] - df['newbalanceOrig'] - df['amount']

# If errorOrig ‚â† 0, it means the sender‚Äôs balance change does not match the transaction amount.
```

```python
df[df['errorOrig'] != 0].shape[0]
```

Out of **6,362,620 total rows**, around **5.87 million transactions (~93%)** have **inconsistent sender balances**.

This means:

- In these rows, the change in the sender's balance doesn't **logically** match the transaction amount.

#### Possible Reasons

1. **System logging delays** ‚Äì Balances might not update in real-time.
2. **Transaction type differences** ‚Äì Not all transactions deduct money (e.g., cash-ins, reversals).
3. **Data anonymization** ‚Äì The dataset creators may have masked or corrupted balance values.
4. **Fraudulent patterns** ‚Äì Many frauds may involve manipulating balance data.

### 2. Does the amount match the change in receiver balance?

```python
df['errorDest'] = df['newbalanceDest'] - df['oldbalanceDest'] - df['amount']
# Again, non-zero values indicate inconsistency in receiver's balance update.

df[df['errorDest'] != 0 ].shape[0]
```

## Fraudsters Prefer Transfer Loops

### 1. Track Linked Transactions

For every fraudulent TRANSFER, check if there's a subsequent CASH_OUT to the same nameDest.

```python
# Find all fraud TRANSFER transactions
fraud_transfers = df[(df['type'] == 'TRANSFER') & (df['isFraud'] == 1)]

# Match with cash out to the same destination
potential_loops = df[
    (df['type'] == 'CASH_OUT') &
    (df['nameOrig'].isin(fraud_transfers['nameDest']))
]

# Merge for analysis
transfer_loops = fraud_transfers.merge(
    potential_loops,
    left_on='nameDest',
    right_on='nameOrig',
    suffixes=('_transfer', '_cashout')
)
```

### 2. Calculate Timing Gap

```python
transfer_loops['time_diff'] = transfer_loops['step_cashout'] - transfer_loops['step_transfer']
```

```python
transfer_loops[['step_transfer','nameOrig_transfer','nameDest_transfer','amount_transfer','step_cashout','nameOrig_cashout', 'nameDest_cashout' ,'amount_cashout','time_diff']].head()
```

### 3. Flag Destination Accounts

The nameDest that frequently:

Receives TRANSFERs

And immediately CASH_OUTs

Are likely mule accounts.

```python
suspicious_mules = transfer_loops['nameDest_transfer'].value_counts()
suspicious_mules
```

## Confusion matrix showing how many frauds were missed by the current system

```python
# Actual and predicted values
y_true = df['isFraud']
y_pred = df['isFlaggedFraud']

# Compute confusion matrix
cm = confusion_matrix(y_true, y_pred, labels=[0, 1])

# Display
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=["Not Fraud", "Fraud"])
disp.plot(cmap="Blues", values_format='d')
plt.title("Confusion Matrix: Flagged vs Actual Fraud")
plt.show()
```

## Heatmap of correlation between numeric variables and isFraud

```python
# Select numeric columns only
numeric_df = df.select_dtypes(include='number')

# Compute correlation matrix with isFraud
correlation = numeric_df.corr()['isFraud'].sort_values(ascending=False)

# Drop isFraud to avoid self-correlation in the plot
correlation = correlation.drop('isFraud')

# Create a heatmap
plt.figure(figsize=(8, 5))
sns.heatmap(correlation.to_frame(), annot=True, cmap='coolwarm', cbar=False)
plt.title('Correlation of Numeric Features with isFraud')
plt.show()
```
